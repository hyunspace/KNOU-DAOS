## 데이터 수집

```python
import requests

api_key = ""
url = f"http://openapi.seoul.go.kr:8088/{api_key}/json/energyUseDataSummaryInfo/1/5/"

# Test Parmas
params = {
    'KEY': api_key,
    'TYPE': 'json',
    'SERVICE': 'energyUseDataSummaryInfo',
    'START_INDEX': 1,
    'END_INDEX': 1000,
    'YEAR': '2025',
    'MON': '05'
}

## 데이터 수집
response = requests.get(url, params=params)

## 호출 성공/실패 출력
print(response.json())

```

## 데이터 프레임 생성

```python
import pandas as pd


## requests로부터 response, body, item 항목 읽기
response = requests.get(url, params=params)
data = response.json()['energyUseDataSummaryInfo']['row']
print(type(data))
print(len(data))
print(data[0])
```

```python
#df.to_csv('air_pollution.csv', index=False, encoding='utf-8')
df = pd.DataFrame(data)

df.to_csv('energy.csv', index=False, encoding='utf-8')

## JSON 형식 데이터 저장
df.to_json('energy.json', indent=4, force_ascii=False)

## EXCEL 형식 데이터 저장
df.to_excel('energy.xlsx', index=False)

## HTML 형식 데이터 저장
energy_html = df.to_html()

with open('energy.html', 'w') as f:
    f.write(energy_html)
```
